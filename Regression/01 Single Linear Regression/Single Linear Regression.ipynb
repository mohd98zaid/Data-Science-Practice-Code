## Simple Linear Regression

```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the dataset
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
#split into dependent and independent variable
X = df.iloc[:,:-1].values
y = df.iloc[:,-1:].values
# X,y
```

```python
#import the libaries and split the dataset in training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .2, random_state=0)
```

```python
#import the linear regression libaries
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
#fit the test set
lin_reg.fit(X, y)
```

```python
#predict the test set result
y_pred = lin_reg.predict(X_test)
```

```python
#visualize the training set
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title("Salary vs Experience(Training sets)")
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
#visualize the test sets
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, lin_reg.predict(X_train), color = 'blue')
plt.title("Salary vs Experence (Test sets)")
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
