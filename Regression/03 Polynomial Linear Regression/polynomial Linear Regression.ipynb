```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the datasets
df = pd.read_csv("Position_Salaries.csv")
df.head()
```

```python
df.info()
```

```python
#split the dataset into dependent and independent variable
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
#Training the datasets
from sklearn.linear_model import LinearRegression
Poly_Lreg = LinearRegression()
Poly_Lreg.fit(X,y)
```

```python
#Training the datasets in polynomial regression
from sklearn.preprocessing import PolynomialFeatures

Poly_reg = PolynomialFeatures(degree=4)
X_poly = Poly_reg.fit_transform(X)
Poly_lin_reg = LinearRegression()
Poly_lin_reg.fit(X_poly, y)
```

```python
#Visualize the result in single Linear regression
plt.scatter(X,y,color='Red')
plt.plot(X, Poly_Lreg.predict(X), color = 'Blue')
plt.xlabel('Salary')
plt.ylabel('Level of experence')
plt.title('Salary v/s Level of Experience(Single Linear)')
plt.show()
```

```python
# Visualize the polynomial regression
plt.scatter(X,y,color='red')
plt.plot(X,Poly_lin_reg.predict(Poly_reg.transform(X)), color='blue')
plt.title('Salary v/s Level of Experience(Polynomial Linear)')
plt.xlabel('Salary')
plt.ylabel('Level of experence')
plt.show()
```

```python
# Visualize the polynomial curve with more smoothness
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X,y , color = 'Red')
plt.plot(X_grid, Poly_lin_reg.predict(Poly_reg.fit_transform(X_grid)), color='Blue')
plt.title('Salary v/s Level of Experience(Polynomial Linear(smooth))')
plt.xlabel('Salary')
plt.ylabel('Level of experence')
plt.show()
```

```python
#Predict the result
Poly_lin_reg.predict(Poly_reg.fit_transform([[6.5]]))
```
